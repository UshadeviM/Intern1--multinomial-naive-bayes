# **Hill and Valley Prediction using Logistic Regression**

# **Objective**

To develop an Machine Learning Model to predict Hill and Valley using Logistic Regression method

# **Data Source**

This dataset was taken from the github library which is maintained at YBI Foundation. Each record represents 100 points on a two dimensional graph. When plotted in order (from 1 through 100) as the Y coordinate, the point will create either a Hill (a "bump" in the terrain) or a Valley (a "Dip" in the terrain).

# **Import Library**

```python
import pandas as pd
```

```python
import numpy as np
```

```python
import matplotlib.pyplot as plt
```

# **Import Data**

```python
df = pd.read_csv("https://github.com/YBI-Foundation/Dataset/raw/main/Hill%20Valley%20Dataset.csv")
```

# **Data Visualization**

# **Define Target Variable (y) and Feature Variables (X)**

```python
df.columns
```

```python
y = df['Class']
```

```python
y.shape
```

```python
y
```

**for defining x variable we can use .drop function instead of mentioning each attribute ('V1', 'V2', 'V3',......)**

```python
x = df.drop('Class', axis = 1)
```

```python
x.shape
```

```python
x
```

# **Train Test Split**

**Scaling data**

```python
from sklearn.preprocessing import StandardScaler
```

```python
ss = StandardScaler()
```

```python
x = ss.fit_transform(x)
```

```python
x
```

```python
pd.DataFrame(x).describe()
```

**After Standardization Mean is Zero and Standard Deviation is One**

```python
from sklearn.model_selection import train_test_split
```

```python
x_train, x_test, y_train, y_test = train_test_split(x, y, train_size = 0.7, random_state= 2529)
```

```python
x_train.shape, x_test.shape, y_train.shape, y_test.shape
```

# **Modeling**

```python
from sklearn.linear_model import LogisticRegression
```

```python
model = LogisticRegression()
```

```python
model.fit (x_train, y_train)
```

# **Model Evaluation**

```python
model.intercept_
```

```python
model.coef_
```

# **Prediction**

```python
y_pred = model.predict(x_test)
```

```python
y_pred
```

# **Model Accuracy**

```python
from sklearn.metrics import confusion_matrix, classification_report, accuracy_score
```

```python
print(confusion_matrix(y_test,y_pred))
```

```python
print(classification_report(y_test, y_pred))
```

```python
accuracy_score(y_test,y_pred)
```
